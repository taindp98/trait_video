```python
def blending(clear, fog):
    """
    Alpha blending
    param:
        clear: is original image path
        fog: is fog image path
    return:
        hazy: is hazy-image
    """
    clear = cv2.imread(clear)
    fog = cv2.imread(fog)
    fog = cv2.resize(fog, clear.shape[1::-1])
    hazy = cv2.addWeighted(clear, 0.5, fog, 0.5, 0)
    return hazy
```
